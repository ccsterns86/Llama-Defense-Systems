```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

# Load your CSV
df = pd.read_csv("averaged_results_w.csv")

# Define colors
colors = {
    1: 'gold',
    2: 'sandybrown',
    3: 'turquoise',
    4: 'blue',
    5: 'orchid'
}
```

# Comparison of parameters over varying number of llamas

```python
# Group by Parameter and Value
for (parameter, value), group in df.groupby(["Parameter", "Value"]):
    plt.figure(figsize=(10,6))
    
    # Plot one line per Number of Predators
    for predator in sorted(group["Num Predators"].unique()):
        # print(colors.predator)
        sub = group[group["Num Predators"] == predator]
        plt.plot(
            sub["Num llamas"],
            sub["Average Sheep Alive"],
            label=f"Predators: {predator}",
            marker='o',
            color=colors[predator]
        )

    plt.title(f"Comparing {parameter} at value of {value}")
    plt.xlabel("Number of Llamas")
    plt.ylabel("Average Sheep Alive")
    plt.legend()
    plt.ylim(80, 101)
    plt.xticks([1, 2, 3, 4, 5])
    plt.grid(True)
    plt.tight_layout()
    
    plt.show()
```

```python
# Group by Parameter and Predator
for (parameter, preds), group in df.groupby(["Parameter", "Num Predators"]):
    plt.figure(figsize=(10,6))
    
    # Plot one line per Number of Values
    for value in sorted(group["Value"].unique()):
        # print(colors.predator)
        sub = group[group["Value"] == value]
        plt.plot(
            sub["Num llamas"],
            sub["Average Sheep Alive"],
            label=f"Value: {value}",
            marker='o',
            # color=colors[int(value)]
        )

    plt.title(f"Comparing {parameter} at {preds} Predators")
    plt.xlabel("Number of Llamas")
    plt.ylabel("Average Sheep Alive")
    plt.legend()
    plt.ylim(80, 101)
    plt.xticks([1, 2, 3, 4, 5])
    plt.grid(True)
    plt.tight_layout()
    
    plt.show()
```

```python
# Group by Parameter and Num llamas
for (parameter, num_llamas), group in df.groupby(["Parameter", "Num llamas"]):
    plt.figure(figsize=(10,6))
    
    # Plot one line per parameter Value (e.g., Cohesion value)
    for value in sorted(group["Value"].unique()):
        sub = group[group["Value"] == value]
        plt.plot(
            sub["Num Predators"],
            sub["Average Sheep Alive"],
            label=f"Value: {value}",
            marker='o',
            # Use a color palette if you want, e.g. tab10
            # color=some_color_map[value]
        )

    plt.title(f"Comparing {parameter} at {num_llamas} Llamas")
    plt.xlabel("Number of Predators")
    plt.ylabel("Average Sheep Alive")
    plt.legend(title="Parameter Value")
    plt.ylim(80, 101)
    plt.xticks([1, 2, 3, 4, 5])
    plt.grid(True)
    plt.tight_layout()
    
    plt.show()
```
